```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(iris, sep=',')
```

```python
attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
df.columns = attributes
```

```python
df.head()
```

###Col Types
- Numerical -> sepal_lenghth, sepal_width, petal_length, petal_width
- Categorical -> species
-mixed -> None

```python
df['species'].value_counts()
```

```python
# shape

df.shape
```

```python
# Scatter plot -> sepal_length - sepal_width

sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

### Univariate Analysis

- By using petal_width we can be sure that length between 0-8 is setosa
- By using petal_length we can be sure that length between 0-2 is setosa

```python
sns.kdeplot(data=df,  x='sepal_length', hue='species')
```

```python
sns.kdeplot(data=df,  x='sepal_width', hue='species')
```

```python
sns.kdeplot(data = df ,x='petal_width', hue='species')
```

```python
sns.kdeplot(data=df, x='petal_length', hue='species')
```
